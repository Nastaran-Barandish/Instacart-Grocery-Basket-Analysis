### Importing Libraries

```python
# Importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

### Importing Data

```python
# Path
path=r"C:\Users\nabar\OneDrive\Documents\03-2025 Instacart Basket Analysis"
# Dataframe
df = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_cust_merge.pkl'))
```

### Q3

```python
# Histogram
hist = df['order_hour_of_day'].plot.hist(bins = 24, color= 'green')
plt.title('Order Placement Time')
plt.xlabel('Hour of The Day')
```

### Analysis of Order Volume by Hour

The histogram illustrates the distribution of Instacart orders throughout the day. The frequency of orders begins increasing around 6 AM, reaching a peak between 10 AM and 3 PM, and then gradually declines into the evening hours. This trend indicates that most orders are placed during the late morning and early afternoon, suggesting peak shopping hours when customers are actively purchasing groceries.

Senior stakeholders can leverage this insight to optimize staffing, inventory management, and promotional strategies during peak hours, ensuring efficient service and improved customer satisfaction.


### Q4

```python
bar = df['loyalty_flag'].value_counts().plot.bar(color =['purple', 'red', 'pink'])
plt.title('Customer Loyalty Distribution')
plt.xlabel('')
plt.ylabel('Count')
```

### Observations:

The bar chart clearly demonstrates that loyal customers contribute the highest number of orders, significantly outpacing regular and new customers.

Loyal customers generate around 15 million orders, suggesting that Instacart's strongest customer base consists of repeat buyers with a high purchase frequency.

Regular customers contribute approximately 10 million orders, showing a sizable group of shoppers who engage consistently but may not be as committed as loyal ones.

New customers account for roughly 8 million orders, indicating potential growth opportunities. Marketing efforts can be focused on converting these new customers into loyal ones through targeted promotions and engagement strategies.

This insight can help the marketing team tailor campaigns to retain loyal shoppers, increase engagement among regular customers, and encourage repeat purchases from new customers.

### Q5

```python
# Sample
sampled_df = df.sample(n=100000, random_state=1)
```

```python
# Line chart
line_price_hod = sns.lineplot(data = sampled_df, x = 'order_hour_of_day', y = 'prices',color='#68bf43')

plt.xlabel("Hour of Day")
plt.ylabel("Average Price of Products Purchased")
plt.title("Average Price of Instacart Products Purchased by Hour of Day")
```

### Observations:

The line chart reveals that expenditure fluctuates depending on the hour of the day. There appear to be distinct peaks in spending around 10 AM and 3 PM, indicating that customers tend to make higher-priced purchases during these hours. Early morning and late evening transactions show lower mean expenditures, suggesting that customers buying at these times may be purchasing smaller or essential items rather than large grocery hauls.

This insight can be valuable for Instacart stakeholders in optimizing marketing strategies, inventory planning, and targeted promotions during peak spending hours.

### Q7

```python
scatter = sns.scatterplot(x='Age', y='income', data=sampled_df)

plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Financial Patterns Across Age Groups')
```

### Relationship Between Age and Income

The scatterplot demonstrates the distribution of income across various ages. Each point represents an individual's income in relation to their age. The visualization indicates that income levels tend to be higher among individuals in their middle-age range (approximately 30 to 60 years old), while younger individuals (below 30) and older individuals (above 60) generally have lower income levels.

This trend suggests that earning potential increases as individuals gain more experience in the workforce and peaks during middle age before gradually declining as they approach retirement. This insight can inform marketing campaigns by helping target specific age groups based on their spending power, ensuring that promotions and offers align with customer financial capabilities.
