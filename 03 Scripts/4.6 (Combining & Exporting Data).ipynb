### Merging Dataframes

Product_id is the key column.

```python
# Test Merge
pd.merge(orders_products_combined , df_prods , on = ['product_id'] , how = 'inner', indicator = 'match' )
```

```python
# Merging 'orders_products_combined' and '_df_pords'
ords_prods_merge = orders_products_combined.merge(df_prods , on = ['product_id'] , indicator = 'match' )
```

```python
# Confirm the results of the merge using the merge flag
ords_prods_merge.drop('_merge', axis = 1 )
```

```python
ords_prods_merge['_merge'].value_counts()
```

The merged dataframe contains 32,404,859 rows, where each row includes data from both input datasets. This result is achieved through the use of an inner join, ensuring that only matching records from both datasets are included.
