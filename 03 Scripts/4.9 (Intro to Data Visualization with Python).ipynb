```python
pivot_table = ords_prods_merge.pivot_table(values='prices', index='order_hour_of_day', aggfunc='sum')
```

```python
# line chart of the pivot table
line_3 = pivot_table.plot(kind='line', figsize=(10, 6), legend=False, color='red')

# Add titles and labels
plt.title('Total Spent by Hour of Day')
plt.xlabel('Hour of The Day')
plt.ylabel('Total Spent')
plt.xticks(range(24))
plt.show()
```

```python
line_3.figure.savefig(os.path.join(path, '04 Analysis','Visualization', 'TotalSpent_HourofDay.png'))
```
